
### üìù Task 1: Shopping Cart System (List)

Write a Python program that simulates a simple shopping cart:

- Store product names in a list
- Allow the user to:
  - Add new products to the cart
  - Remove a product from the cart
  - View all products in the cart
- Display the total number of items in the cart

```python
"""
Shopping Cart System

This program allows the user to:
1. Add items to a shopping cart
2. Remove items from the cart
3. View all items currently in the cart
4. Exit the program

At the end, it displays:
- Total number of items in the cart
- Final list of items

The cart is stored using a Python list.
"""

# Create an empty list to store cart items
cart = []

# Run the program continuously until the user chooses to exit
while True:
    # Display menu and take user choice
    c = int(input(
        "Enter option:\n"
        "1 to add to cart\n"
        "2 to remove from cart\n"
        "3 to check cart\n"
        "4 to exit\n"
    ))

    # Option 1: Add items to cart
    if c == 1:
        n = int(input("How many items? "))
        for i in range(n):
            ent = input("Enter the item: ")
            cart.append(ent)   # Add item to cart
        print("Items added to cart:", cart)

    # Option 2: Remove item from cart
    elif c == 2:
        print("Current cart:", cart)
        ent = input("Enter the item to remove: ")
        if ent in cart:
            cart.remove(ent)   # Remove item if found
            print("Updated cart:", cart)
        else:
            print("Item not found!")

    # Option 3: View cart items
    elif c == 3:
        print("The products in your cart are:", cart)

    # Option 4: Exit the loop
    elif c == 4:
        print("Exit done!")
        break

    # Handle invalid menu choices
    else:
        print("Invalid option!")

# Final cart summary after exiting the loop
print("The total number of items in your final cart is:", len(cart))
print("The items in your final cart are:", cart)
```

### üìù Task 2: Student Attendance System (Set)

Write a Python program that manages student attendance:

- Store roll numbers of students present today in a set
- Store roll numbers of students present yesterday in another set
- Display:
  - Students present on both days
  - Students present only today
  - Students absent today but present yesterday


```python
"""
Student Attendance System

This program uses sets to:
- Store roll numbers of students present today and yesterday
- Find students present on both days
- Find students present only today
- Find students absent today but present yesterday
"""

# Roll numbers of students present today
rolltoday = {1, 2, 3, 4, 5, 9}

# Roll numbers of students present yesterday
rollyest = {1, 3, 4, 5, 6, 7}

# Students present on both days
print("Present on both days:", rolltoday & rollyest)

# Students present only today
print("Present only today:", rolltoday - rollyest)

# Students absent today but present yesterday
print("Absent today but present yesterday:", rollyest - rolltoday)
```

### üìù Task 4: Exam Marks Analysis (List + Tuple)

Write a Python program that:

- Stores marks of students in a list
- Calculates:
  - Highest mark
  - Lowest mark
  - Average mark
- Stores the calculated results in a tuple
- Displays all values clearly


```python
"""
Exam Marks Analysis System

This program:
- Stores marks of 3 students using lists
- Finds highest, lowest, and average marks
- Stores the calculated results in tuples
- Displays all details clearly

Concepts used:
- List
- max(), min()
- len()
- Tuple
"""

# Marks of three students
s1 = [60, 55, 64, 51, 68]
s2 = [70, 61, 57, 72, 80]
s3 = [51, 23, 44, 55, 60]


# student 1: details
print("Details of student 1:")

# Highest marks
hm1 = max(s1)
print("Highest marks:", hm1)

# Lowest marks
lm1 = min(s1)
print("Lowest marks:", lm1)

# Calculating total marks manually
tot1 = s1[0] + s1[1] + s1[2] + s1[3] + s1[4]

# Calculating average
avg1 = tot1 / len(s1)
print("Average marks:", avg1)
print("\n")


# Student 2 
print("Details of student 2:")

hm2 = max(s2)
print("Highest marks:", hm2)

lm2 = min(s2)
print("Lowest marks:", lm2)

tot2 = s2[0] + s2[1] + s2[2] + s2[3] + s2[4]
avg2 = tot2 / len(s2)
print("Average marks:", avg2)
print("\n")


# Student 3 
print("Details of student 3:")

hm3 = max(s3)
print("Highest marks:", hm3)

lm3 = min(s3)
print("Lowest marks:", lm3)

tot3 = s3[0] + s3[1] + s3[2] + s3[3] + s3[4]
avg3 = tot3 / len(s3)
print("Average marks:", avg3)
print("\n")


# Storing results in tuples
print("Highest, Lowest and Average marks of each student (stored in tuple):")

t1 = (hm1,lm1,avg1)
print(t1)

t2 = (hm2,lm2,avg2)
print(t2)

t3 = (hm3,lm3,avg3)
print(t3)
```

### üìù Task 5: Duplicate Data Cleaner (List ‚Üí Set)

Write a Python program that:

- Takes a list of email addresses as input
- Removes duplicate email addresses using a set
- Converts the cleaned data back into a list
- Displays both original and cleaned lists


```python
"""
Duplicate Email Cleaner

This program:
- Takes multiple email inputs from the user
- Stores them in a list
- Removes duplicate emails using a set
- Converts the set back into a list
- Displays original and cleaned email lists

Concepts used:
- List
- Set
- for loop
- Type conversion
"""

# Taking number of emails from user
n = int(input("Enter the number of emails you'd like to input: "))

# Empty list to store emails
emails = []

# Taking email inputs
for i in range(n):
    em = input(f"Enter email {i+1}: ")
    emails.append(em)

# Display original email list
print("Original emails:\n", emails)

# Removing duplicates using set and converting back to list
cln_ems = list(set(emails))

# Display cleaned email list
print("Emails after removing duplicates:\n", cln_ems)
```

## Strings in Python

### What is a String?
A string is a sequence of **characters** used to store and manipulate text.

Key points:
- Strings are **immutable**
- Enclosed in single `' '` or double `" "` quotes
- Very common in data cleaning and preprocessing

Examples:
- Names
- Emails
- Text data


```python
# Create a string
str1="sama";

# Print the string
print(str1)
print(str1[0])

# Check the type of the string
print(type(str1))
```

## String Operations

Python provides various operations to work with strings:
- Concatenation
- Repetition
- Indexing
- Length calculation


```python
# Create two strings
str0="samuu"
str1="sama"
print(str0,str1)

# Concatenate strings
str01=str0+""+str1
print(str01)

# Repeat a string
print(str1*5)

# Access a character using index
print(str0[3])

# Find length of string
print(len(str0))
```

## Common String Methods

Some frequently used string methods:
- `lower()`
- `upper()`
- `strip()`
- `replace()`
- `split()`
- `find()`

These methods are widely used in **text preprocessing**.


```python
# Create a string
str2="  Python Programming  ,Language  " #how to remove unneccessary space(spaces are characters too in python)
print(len(str2))
# Convert to lowercase
print(str2.lower())

# Convert to uppercase
print(str2.upper())

# Remove whitespace
print(str2.strip())
print(len(str2.strip()))

# Replace characters
print(str2.replace("Python","Java"))

# Split the string
print(str2.split())
print(str2.split(","))

# Find a substring
print(str2.find("Python"))
print(str2.find("Python",0,3))
print(str2.find("Programming",0,3)) 
  #index of "programming" is 4 but its asked to search only betn 0 to 3
print(str2.find("c++")) 
  #if not found then returns -1
```

## String Slicing

String slicing allows extracting a portion of a string.

Syntax:

string[start : end]


Used for:
- Extracting substrings
- Text processing


```python
# Create a string
str3="Python Programming Language"

# Slice first 5 characters
print(str3[0:3])
print(len(str3))

# Slice last 5 characters
print(str3[-5:])

# Slice middle part
print(str3[])

```

## Regular Expressions (Regex)

Regular Expressions are used to **search, match, and manipulate text**.

They are very powerful for:
- Pattern matching
- Data validation
- Text extraction

Python provides the `re` module for regex operations.


```python
import re
```

## Common Regex Functions

- `re.search()` → Searches for a pattern
- `re.findall()` → Returns all matches
- `re.sub()` → Replaces matches


```python
# Define a text string
text="my phone no is 9768334455 and secondary no is 9818334543" 

# shows if the input matches the required searched item like 10 digits in a number here 

# Search for a pattern
search_result=re.search(r"\d{10}",text)
print(search_result)

# Find all matches
search_result=re.findall(r"\d{10}",text)
print(search_result)


# Replace using 
masked_text=re.sub(r"\d","*",text)
print(masked_text)

```

```python
# Define a text string
text1="my phone no is 976833445 and secondary no is 9818334543" 
#first no has 9 numbers only in text1
# Search for a pattern
search_result1=re.search(r"\d{10}",text1)
print(search_result1)

# Find all matches
search_result1=re.findall(r"\d{10}",text1)
print(search_result1)

# Replace using 
masked_text1=re.sub(r"\d","*",text1)
print(masked_text1)
```

```python
text_search=re.findall(r"[a-zA-Z][a-z]+",text)
print(text_search)

text_search=re.findall(r"[a-zA-Z][a-z ]+",text)
print(text_search)
```

## Why File Handling is Important?

File handling allows a program to:
- Store data permanently
- Read data from files
- Write results to files

In real-world applications, file handling is used for:
- Logging
- Data storage
- Reading datasets
- Report generation


## Types of File Operations

Python mainly supports the following file operations:
- Read (`r`)
- Write (`w`)
- Append (`a`)

Files are handled using the built-in `open()` function.


## Opening a File

Basic syntax:

open("filename", "mode")



Common modes:

`r` - Read - Default value. Opens a file for reading, error if the file does not exist

`a` - Append - Opens a file for appending, creates the file if it does not exist

`w` - Write - Opens a file for writing, creates the file if it does not exist

`x` - Create - Creates the specified file, returns an error if the file exists

```python
# Open a file in read mode
file=open("student.txt","w")

# Open a file in write mode
file.write("name:ram\n")
# Open a file in append mode
file.writelines(["this is first line \n","this is second line"])

#close file
file.close()
```

Read mode (`r`) is used to:
- Read existing file content

Common methods:
- `read()` → reads entire file
- `readline()` → reads one line
- `readlines()` → reads all lines as a list


```python
# Open file in read mode
file=open("student.txt","r")

# Read entire content
print(file.read())

# Read one line
print(file.readline())

# Read all lines


# Close the file
```

## Appending to a File

Append mode (`a`) is used to:
- Add new content at the end of a file
- Preserve existing data


```python
# Open file in append mode
file=open("student.txt","a")

# Append content to the file
file.write("\n this is new appended line")
# Close the file
file.close()
```

## Using with Statement

The `with` statement automatically:
- Opens the file
- Closes the file after use

This is the **recommended approach**.


```python
# Use with statement to write to a file
with open("student.txt","a")as f:
    f.write("\n this is with open file")
# Use with statement to read from a file
with open("student.txt","r")as f:
    print(f.read())
```
